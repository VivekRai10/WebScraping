***Q1)Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.***

```python

import pandas as pd
from bs4 import BeautifulSoup
import requests
from lxml import html

wikipageContent=requests.get(
     'https://en.wikipedia.org/wiki/Main_Page'
)

tree = html.fromstring(wikipageContent.content)

_to_scrape = ['h1','h2','h3']
l1,l2,l3 = [],[],[]
while True:

    for element in _to_scrape:
        path = f'//{element}'
        for j in range(len(tree.xpath(path))):
            if element == 'h1':
                l1.append(tree.xpath(path)[j].text_content())
            elif element == 'h2':
                l2.append(tree.xpath(path)[j].text_content())
            else:
                custom_path = path + '/span'
                l3.append(tree.xpath(custom_path)[j-1].text_content())  
    break

dict = {'h1':l1,'h2':l2,'h3':l3}
wiki_header_df = pd.DataFrame.from_dict(dict, orient='index').T
wiki_header_df
```

***Q2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.***

```python
import urllib.request
url = 'https://www.imdb.com/list/ls091520106/'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36'
}

sauce = urllib.request.urlopen(url).read()

soup = BeautifulSoup(sauce,'html.parser')


blocks = soup.find_all('h3',attrs={'class':'lister-item-header'})          
rating = soup.find_all('div',attrs={'class':'ipl-rating-widget'})   

assert len(blocks) == len(rating)                                   

moviename,movieyear,movierating = [],[],[]                                         

for block in blocks:
    moviename.append(block.a.text)                                                            
    movieyear.append(block.find('span', class_ = 'lister-item-year text-muted unbold').text)  

    
for r in rating:
    movierating.append(r.find('span', class_ = 'ipl-rating-star__rating').text)       
    
imdb_top100_df = pd.DataFrame({'MovieName':moviename,'MovieYear':movieyear,'MovieImdbRating':movierating}) 

imdb_top100_df.to_csv('imdb_top100_movies.csv',index=False)

imdb_top100_df
```

***Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’***

```python
URL = 'https://bookpage.com/reviews'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')

Book_Name=[]
Author_Name=[]
Genre=[]
Book_Review=[]

name=soup.find_all('h4', class_="italic")
for x in name:
    for y in x.find_all('a'):
        Book_Name.append(y.text)
        
for i in soup.find_all('p', class_="sans bold"):
        Author_Name.append(i.text)
        
for z in soup.find_all('p',class_="genre-links hidden-phone"):
    for k in z.find_all('a'):
        Genre.append(k.text) 
        
for o in soup.find_all('p',class_="excerpt"):
        Book_Review.append(o.text)  

df= pd.DataFrame({'Book Name': Book_Name[0:5], 'Author Name': Author_Name[0:5], 'Genre': Genre[0:5],'Book Review': Book_Review[0:5]})

df
```

***Q8) Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.***

```python
import re

forcastpageContent=requests.get(
     'https://forecast.weather.gov/MapClick.php?lat=37.77493000000004&lon=-122.41941999999995#.X7vZI1UzbIU')

tree = html.fromstring(forcastpageContent.content)

period = []
shrt_des = []
temp_f = []
des = []

for t in tree.xpath("//div[contains(@class,'row-forecast')]"):
    
    if 'ight' in t.text_content():
        temp = t.text_content().replace('ight','ight ')
        period.append(temp.split()[0])
        shrt_des.append(temp.split(',')[0])
        temp_f.append(int(''.join(re.findall('[0-9]',temp))[:2]))
        des.append(''.join(temp.split(' ',maxsplit=1)[1:]))
        
    else:
        temp = t.text_content().replace('day','day ')
        period.append(temp.split()[0])
        shrt_des.append(temp.split(',')[0])
        temp_f.append(int(''.join(re.findall('[0-9]',temp))[:2]))
        des.append(''.join(temp.split(' ',maxsplit=1)[1:]))
        
sf_temp_df = pd.DataFrame({'period':period,'short_des':shrt_des,'temperature':temp_f,'description':des})

sf_temp_df
```

***Q9) Write a python program to scrape ‘software developer’ job listings from ‘Monster.com’. It should include all the jobs listed for the next 5 pages.***

```python

page_num = 0
index = 2
job_titles, job_company = [],[]

job_location, job_exp_req, job_sal = [],[],[]

while page_num < 5:
    
    if page_num == 0:
        url = 'https://www.monsterindia.com/search/software-developer-jobs'
    else:
        url = f'https://www.monsterindia.com/search/software-developer-jobs-{index}?searchId=55527db6-ebcc-4c07-a7d1-229cbacde155'

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36'
    }

    sauce = urllib.request.urlopen(url).read()

    soup = BeautifulSoup(sauce,'html.parser')
    
    
    title_block = soup.find_all('h3',attrs={'class':'medium'})
    company_block = soup.find_all('span',attrs={'class':'company-name'})
    detail_block = soup.find_all('span',attrs={'class':'loc'})
    
    for title in title_block:
        #print(re.sub(' |\n','',title.text))
        job_titles.append(re.sub(' |\n','',title.text))
        
    for company in company_block:
        #print(re.sub(' |\n','',company.text))
        job_company.append(re.sub(' |\n','',company.text))
        
        
    for i,value in enumerate(detail_block):
        
        if i%3 == 0:
            job_location.append(re.sub(' |\n','',value.text))
            #print(re.sub(' |\n','',value.text))            
        if i%3 == 1:
            job_exp_req.append(re.sub(' |\n','',value.text))
            #print(re.sub(' |\n','',value.text))
        if i%3 ==2:
            job_sal.append(re.sub(' |\n','',value.text))
            #print(re.sub(' |\n','',value.text))        
    
    page_num += 1
    index += 1
    

df = pd.DataFrame({'job_title':job_titles,'job_company':job_company,'job_locations':job_location,'experience_req':job_exp_req,'job_salary':job_sal})

df.to_csv('monster_swd_jobs.csv',index=False)

df
```

***Q10) Write a python program to scrape ‘data scientist’ job listings for location ‘New Delhi’ from ‘Monster.com’. It should include all the jobs listed for the next 5 pages.***

```python


import urllib.request
import pandas as pd
import requests

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

opts = Options()
driver = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=opts)
opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36")


page_num = 0
index = 25
job_titles, job_company = [],[]

job_location, job_exp_req, job_sal = [],[],[]

while page_num < 5:
    
    
    if page_num == 0:
        url = 'https://www.monsterindia.com/srp/results?query=Data%20Scientist%20&locations=Delhi&searchId=4c6b7342-ac00-4921-ad60-1c08f09abcc8'
    else:
        url = f'https://www.monsterindia.com/srp/results?start={index}&sort=1&limit=25&query=Data%20Scientist%20&locations=Delhi&searchId=4c6b7342-ac00-4921-ad60-1c08f09abcc8'
        index += 25        
    

    driver.get(url)
    
    blocks = driver.find_elements_by_xpath("//div[@class='card-body card-body-apply pd10']")
    
    
    for block in blocks:
        job_titles.append(block.find_element_by_xpath('.//a').text)
        
        job_company.append(block.find_element_by_xpath('.//span[@class="company-name"]').text)
        
        job_location.append(block.find_element_by_xpath('.//div[@class="col-xxs-12 col-sm-5 text-ellipsis"]').text)
        
        job_exp_req.append(block.find_element_by_xpath('.//div[@class="exp col-xxs-12 col-sm-3 text-ellipsis"]').text)
        
        job_sal.append(block.find_element_by_xpath('.//div[@class="package col-xxs-12 col-sm-4 text-ellipsis"]').text)
        
    page_num += 1
    

df_delhi = pd.DataFrame({'job_title':job_titles,'job_company':job_company,'job_locations':job_location,'experience_req':job_exp_req,'job_salary':job_sal})

df_delhi.to_csv('monster_swd_jobs_delhi.csv',index=False)

df_delhi
```
